# Question 1

Using the dataset "us-cities-top-1k-multi-year.csv":
1. Create a line graph of total population over time for the states: NY, PA, CT, NJ (10 pts)
2. Be sure to include a title of the graph, labels on your axis, and a legend. (5 pts)
3. Specify colors to be used in your code. (5 pts)

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/Users/shelsea/Downloads/us-cities-top-1k-multi-year.csv")
state_map = {"New York": "NY", "Pennsylvania": "PA", "Connecticut": "CT", "New Jersey": "NJ"}
df_filtered = df[df["State"].isin(state_map.keys())].copy()
df_filtered["StateCode"] = df_filtered["State"].map(state_map)
pivot = df_filtered.groupby(["year", "StateCode"])["Population"].sum().unstack()

pivot.plot(figsize=(10, 6), marker="o", color=["#2ca02c", "#ff7f0e", "#1f77b4", "#d62728"])
plt.title("Total City Population Over Time: NY, PA, CT & NJ")
plt.xlabel("Year")
plt.ylabel("Total Population")
plt.legend(title="State")
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.show()
```
